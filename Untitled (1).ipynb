```python
import numpy as np

import tensorflow as tf
import tensorflow_hub as hub
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("root2ai - Data.csv")

df = df.dropna()
```

```python
len(df['Target'].unique())
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df['Target'])
y = le.transform(df['Target'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['Text'],y , test_size=0.33, random_state=42)
# X_train = list(X_train)
# y_train = list(y_train)
# X_test = list(X_test)
# y_test = list(y_test)
```

```python
model1 = "https://tfhub.dev/google/nnlm-en-dim128-with-normalization/2"
hub_layer = hub.KerasLayer(model1, input_shape=[], dtype=tf.string, trainable=True)
hub_layer(X_train)
```

```python
model = tf.keras.Sequential()
model.add(hub_layer)
# model.add(tf.keras.layers.Dense(50,input_shape=(1,50), activation='relu'))
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(11,activation = 'softmax'))

model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=[tf.metrics.SparseCategoricalAccuracy(name='accuracy')])
model.fit(X_train, y_train, epochs=3, batch_size = 32)
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
model.save("model.h5")
```

```python
yt = model.predict(X_test)
y_pred = []
for c in yt:
  y_pred.append(np.argmax(c))
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
